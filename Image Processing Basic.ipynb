## 1) RGB to HSV and Vice Versa

```python
from skimage import color
from skimage import data
from pylab import * 
```

#### Convert RGB to HSV

```python
img_hsv = color.rgb2hsv(img)
```

### Convert back to RGB

```python
img_rgb = color.hsv2rgb(img_hsv)
```

```python
figure(0)
```

```python
io.imshow(img_hsv)
```

```python
figure(1)
```

```python
io.imshow(img_rgb)
```

# 2) RGB to XYZ and Vice Versa

```python
img = io.imread("photo-1460353581641-37baddab0fa2.jpg")
```

### Convert to XYZ

```python
img_xyz = color.rgb2xyz(img)
```

### Covert back to RGB

```python
img_rgb = color.xyz2rgb(img_xyz)
```

```python
figure(0)
```

```python
io.imshow(img_xyz)
```

```python
figure(1)
```

```python
io.imshow(img_rgb)
```

# 3) RGB to LAB and Vice Versa

### Convert RGB to LAB

```python
img_lab = color.rgb2lab(img)
```

```python
img_rgb = color.lab2rgb(img_lab)
```

```python
io.imshow(img_lab)
```

```python
io.imshow(img_rgb)
```
